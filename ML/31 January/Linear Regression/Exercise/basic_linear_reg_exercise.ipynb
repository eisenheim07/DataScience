```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.tail()
```

```python
plt.figure(figsize=(5,3))
plt.scatter(df.year, df['per capita income (US$)'], marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
```

```python
model = LinearRegression()
model.fit(df[['year']], df['per capita income (US$)'])
```

```python
model.score(df[['year']], df['per capita income (US$)'])
```

```python
model.predict([[2015]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(5,3))
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year, df['per capita income (US$)'], marker='+', color='red')
plt.plot(df.year, model.predict(df[['year']]), color='blue')
```

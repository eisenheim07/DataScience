```python
import numpy as np
```

```python
def gradient_descent(x, y):
    mcurr = bcurr = 0
    n = len(x)
    learning_rate = 0.08
    
    for i in range(150):
        y_pred = mcurr*x + bcurr
        
        cost = (1/n)*sum([val**2 for val in (y-y_pred)])
        
        dm = (-2/n)*sum((y - y_pred)*x)
        db = (-2/n)*sum(y - y_pred)
        
        mcurr = mcurr - learning_rate*dm
        bcurr = bcurr - learning_rate*db
        
        print(f'COST : {cost}, M : {mcurr}, B : {bcurr}, ITER : {i}')

x = np.array([1,2,3,4,5])
y = np.array([10, 20, 30, 40, 50])

gradient_descent(x, y)
```

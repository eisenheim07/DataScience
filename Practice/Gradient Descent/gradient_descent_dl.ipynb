```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def mse(X, y):
    return 1/len(X)*(sum([val**2 for val in abs(X - y)]))
```

```python
X = np.array([11,2,3,4,5])
y = np.array([11,12,13,14,15])
mse(X, y)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradientDescent(X, y, y_true):
    w1 = w2 = 1
    bias = 0
    n = len(X)
    learning_rate = 0.0001
    
    for i in range(1000):
        equ = X*w1 + y*w2 + bias
        y_pred = sigmoid(equ)
        
#         cost = mse(y_true, y_pred)
        cost = log_loss(y_true, y_pred)
        
        w1d = 1/n*(sum(X*(y_pred-y_true)))
        w2d = 1/n*(sum(y*(y_pred-y_true)))
        bias_d = (1/n)*sum(y_pred-y_true)
        
        w1 = w1 - learning_rate*w1d
        w2 = w2 - learning_rate*w2d
        bias = bias - learning_rate*bias_d
        
        print(f'EPOCH: {i}, COST: {cost}, W1: {w1}, W2: {w2}, BIAS: {bias}')
```

```python
gradientDescent(df.age, df.affordibility, df.bought_insurance)
```

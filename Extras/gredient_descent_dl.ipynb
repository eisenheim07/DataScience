```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

# print("Y Predicted = ", sigmoid(23))
```

```python
# typical way to find MEAN ABSOLUTE ERROR
def meanAbsError(x, y):
    values = []
    loss = []
    for i in range(len(x)):
        values.append(sigmoid(x[i]))
#     print("Y PREDICTED = ", values)
        
    for i in range(len(values)):
        loss.append(abs(y[i] - values[i]))
#     print("LOSSES = ", loss)
    
    mae = 0
    for i in range(len(loss)):
        mae += loss[i]
#     print("MEAN ABSOLUTE ERROR = ", abs(mae/len(loss)))
    
meanAbsError([1, 100, 33, 43, 54, 67], [0, 1, 0, 1, 0, 1])
```

```python
# numpy way to find MEAN ABSOLUTE ERROR
y_predicted = [1, 1, 0, 0, 1]
y_true = [0.30, 0.7, 1, 0, 0.5]

def mae(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        print(yt, " " , yp, " " , abs(yt-yp))
        total_error += abs(yt-yp)
#     print()
#     print("TOTAL ERROR = ", total_error)
#     print("MEAN ABSOLUTE ERROR = ", total_error/len(y_true))
        
mae(y_true, y_predicted)
```

```python
# typical way to find MEAN SQUARED ERROR
def meanSquError(x, y):
    values = []
    loss = []
    for i in range(len(x)):
        values.append(sigmoid(x[i]))
#     print("Y PREDICTED = ", values)
    
    for i in range(len(values)):
        loss.append((y[i] - values[i])*(y[i] - values[i]))
#     print("LOSSES = ", loss)
    
    mae = 0
    for i in range(len(loss)):
        mae += loss[i]
#     print("MEAN SQUARED ERROR = ", mae/len(loss))
    
meanSquError([1, 100, 33, 43, 54, 67], [0, 1, 0, 1, 0, 1])
```

```python
# numpy way to find MEAN SQUARED ERROR
y_predicted = [1, 1, 0, 0, 1]
y_true = [0.5, 0.1, 0.3, 0.9, 0.6]

def mse(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
#         print(yt, " " , yp, " " , yt-yp)
        total_error += (yt-yp)*(yt-yp)
        return total_error
#     print()
#     print("TOTAL ERROR = ", total_error)
#     print("MEAN SQUARED ERROR = ", total_error/len(y_true))
        
mse(y_true, y_predicted)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gredientDescent(age, affordability, y_true, epochs):
    w1 = w2 = 1
    bias = 0
    learning_rate = 0.0001
    n = len(age)
    for i in range(epochs):
        equ = w1*age + w2* affordability + bias
        y_pred = sigmoid(equ)
        cost = log_loss(y_true, y_pred)
        
        dw1 = 1/n*(sum(age*(y_pred-y_true)))
        dw2 = 1/n*(sum(affordability*(y_pred-y_true)))
        
        w1 = w1 - learning_rate*dw1
        w2 = w2 - learning_rate*dw2
        bias = np.mean(y_pred-y_true)
        
        print(f'Epoch: {i}, cost: {cost}, w1: {w1}, w2: {w2}, bias: {bias}')

age = np.array([10, 20, 22, 23, 44, 42, 54, 64, 72, 98])
affordability = np.array([1, 1, 1, 1, 0, 0, 1, 0, 1, 0])
y_true = np.array([1, 0, 0, 0, 1, 0, 1, 0, 1, 1])
gredientDescent(age, affordability, y_true, 5000)
```

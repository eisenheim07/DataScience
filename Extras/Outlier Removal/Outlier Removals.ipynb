```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('heights.csv')
df.head()
```

```python
df.shape
```

```python
sn.histplot(df.height, kde=True)
```

# REMOVAL 3 LEVEL STANDARD DEVIATION

```python
df.height.mean()-3*df.height.std()
```

```python
df.height.mean()+3*df.height.std()
```

```python
df1 = df[(df.height > (df.height.mean()-3*df.height.std())) & (df.height < (df.height.mean()+3*df.height.std()))]
df1.head()
```

```python
sn.histplot(df1.height, kde=True)
```

# REMOVAL 2 LEVEL STANDARD DEVIATION

```python
df2 = df[(df.height > (df.height.mean()-2*df.height.std())) & (df.height < (df.height.mean()+2*df.height.std()))]
df2.head()
```

```python
sn.histplot(df2.height, kde=True)
```
